2.Analiza datelor

```python
print("===== Date lipsă ===== ") 
cl.isna().sum()
```

```python
# Se verifică dacă există rânduri duplicate în dataset
cl.duplicated().sum()
```

```python
# ------------------------------
# Detreminarea valorilor extreme
# ------------------------------

# Selectarea coloanelor numerice
num = cl.select_dtypes(include=['number'])

# Calcularea percentilei de 5% (Q1) și 95% (Q3)
Q1 = num.quantile(0.05)
Q3 = num.quantile(0.95)

# Calcularea IQR: diferența dintre Q3 și Q1
IQR = Q3 - Q1

# Sortarea coloanelor după IQR, descrescător
IQR.sort_values(ascending=False)
```

```python
# 1. Barplot pentru valori lipsa
# Calculam valorile lipsa pe fiecare coloană
missing = cl.isna().sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=missing.index, y=missing.values, color="skyblue") 
plt.xticks(rotation=45)
plt.title("Valori lipsă pe coloane", fontsize=14)
plt.ylabel("Numar valori lipsa")

# Adăugăm etichete pe bare
for i, v in enumerate(missing.values):
    plt.text(i, v + 5, str(v), ha='center', fontsize=10)

plt.show()
```

```python
# ----------------------------
# Eliminarea valorilor extreme
# ----------------------------

# Se păstrează doar rândurile unde toate valorile numerice se află în intervalul [Q1 - 1.5*IQR, Q3 + 1.5*IQR]
cl = cl[
    ((num >= (Q1 - 1.5 * IQR)) & (num <= (Q3 + 1.5 * IQR))).all(axis=1)
]
```

```python
#Stergem coloanele irelevante
cl = cl.drop(['PassengerId', 'Cabin', 'Name','Ticket'], axis=1)
```

3.Procesarea si pregatirea pt modelare

```python
# Selectarea coloanelor numerice 
numerical = cl.select_dtypes(include=['number']).columns.tolist()

# Selectarea coloanele categorice (de tip obiect) 
categorical = cl.select_dtypes(include=['object']).columns.tolist()

# Afișarea rezultatelor
print('Coloane numerice:', numerical)
print('Coloane categorice:', categorical)
```

```python
# Transformarea coloanelor categorice în variabile 'dummy'
cl_final = pd.get_dummies(
    cl,
    columns=['Sex', 'Embarked'],   # coloanele categoriale relevante
    drop_first=True                # evităm colinearitatea
)

# Verificarea primelor 5 rânduri
cl_final.head()
```

```python
X = cl_final.drop('Survived',axis=1) 
y = cl_final['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=21)
```

```python
# --------------------------
# Scalarea dataset-ului folosind MinMaxScaler
# --------------------------

from sklearn.preprocessing import MinMaxScaler

# Crearea obiectului scaler
scaler = MinMaxScaler()

# Învățarea min și max pe baza întregului set X
scaler.fit(X)

# Aplicare scaler pe setul de antrenament
X_train = scaler.fit_transform(X_train)

# Aplicare scaler pe setul de test
X_test = scaler.transform(X_test)

# Transformarea rezultatul în DataFrame pentru lizibilitate, păstrând numele coloanelor
X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)

# Afișare primelor 5 rânduri din seturile scalate
print('========== Antrenare ==========\n')
X_train.head()
```

```python
print(cl_final.shape)
```

```python
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
print('========== Testare ==========\n')
X_test.head()
```

Descrierea variabilelor și a seturilor de date

Datele utilizate în acest proiect includ un total de 8 variabile, dintre care 6 sunt numerice și 2 sunt categoriale(initial,apoi le-am facut si pe cele categoriale tot numerice). Variabilele numerice sunt: Survived, Pclass, Age, SibSp, Parch și Fare. Variabilele categoriale sunt: Sex și Embarked.

Setul de date a fost împărțit în două subseturi: unul pentru antrenare și unul pentru testare. Setul de antrenare conține 495 observații și 8 caracteristici, împreună cu 495 valori țintă. Setul de testare include 213 observații, având aceleași 8 caracteristici, și 213 valori țintă. Această împărțire permite evaluarea corectă a performanței modelului pe date noi.

4.Modelul de regresie Naive Bayes

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score
# -----------------------------
# Gaussian Naive Bayes
# -----------------------------

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, f1_score, classification_report

# Crearea modelului Gaussian Naive Bayes
gaussian = GaussianNB()

# Antrenarea modelului pe datele de training
gaussian.fit(X_train, y_train)

# Prezicerea valorilor pentru setul de test
y_pred_NB = gaussian.predict(X_test)

# Calcularea acurateții pe setul de training
gaussian_train = round(gaussian.score(X_train, y_train) * 100, 2)

# Calcularea acurateții pe setul de test
gaussian_accuracy = round(accuracy_score(y_pred_NB, y_test) * 100, 2)

# Calcularea scorului F1 pe setul de test
gaussian_f1 = round(f1_score(y_pred_NB, y_test) * 100, 2)

# Afișarea rezultatelor
print("Acuratețea pe setul de training :", gaussian_train, "%")
print("Acuratețea modelului           :", gaussian_accuracy, "%")
print("Scorul F1                      :", gaussian_f1, "%")
print("\033[1m--------------------------------------------------------\033[0m")

# Raport de clasificare (precizie, recall, F1 pentru fiecare clasă)
print("Raport de clasificare:\n", classification_report(y_test, y_pred_NB))
```

```python
# -----------------------------
# Matricea de confuzie - Gaussian Naive Bayes
# -----------------------------

plt.rcParams['figure.figsize'] = (10, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 

# Calculare matrice de confuzie
cm = confusion_matrix(y_test, y_pred_NB)

# Transformarea matricei într-un DataFrame pentru vizualizare mai clară
cm_df = pd.DataFrame(cm)

# Vizualizare cu Seaborn

sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title("Matricea de Confuzie - Gaussian Naive Bayes")
plt.ylabel("Valori reale")
plt.xlabel("Valori prezise")
plt.show()
```

```python
from sklearn.metrics import RocCurveDisplay
# -----------------------------
# Curba ROC - Gaussian Naive Bayes
# -----------------------------

plt.rcParams['figure.figsize'] = (14, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 

# Crearea afișajului curbei ROC pe baza estimatorului (modelului)
RocCurveDisplay.from_estimator(
    gaussian,  # modelul Gaussian Naive Bayes
    X_test,    # datele de test
    y_test     # valorile reale
)

# Titlul graficului
plt.title("Curba ROC - Gaussian Naive Bayes")

# Afișare figură
plt.show()

```

Modelul Gaussian Naive Bayes obține o acuratețe de aproximativ 79.81% pe setul de testare, ceea ce indică faptul că în aproape 8 din 10 cazuri modelul prezice corect dacă un pasager a supraviețuit sau nu. În plus, scorul F1 este de 73.62%, un indicator important pentru clasificarea binară, deoarece echilibrează atât precizia, cât și recall-ul. Acest rezultat arată că modelul are un comportament rezonabil în detectarea ambelor clase, chiar dacă performează ceva mai bine pentru clasa majoritară (nesupraviețuitori). Per ansamblu, valorile arată un model cu performanță moderată, adecvat pentru datele relativ simple și limitate ca volum din setul Titanic.


Matricea de confuzie oferă o imagine clară asupra modului în care modelul face predicții pentru fiecare clasă. În acest caz, modelul a prezis corect 110 pasageri care nu au supraviețuit și 60 de pasageri care au supraviețuit, dar a greșit în 18 cazuri etichetând supraviețuitori ca nesupraviețuitori (false negative) și în 25 de cazuri etichetând nesupraviețuitori ca supraviețuitori (false positive). Aceasta arată că modelul tinde să fie puțin mai precaut în a prezice supraviețuirea, ratând o parte dintre supraviețuitori reali. Curba ROC, împreună cu valoarea AUC de 0.85, indică o capacitate bună a modelului de a distinge între cele două clase. O valoare AUC de 0.85 arată că modelul clasifică într-un mod semnificativ mai bun decât întâmplarea și că are un echilibru bun între rata de adevărate pozitive și rata de false pozitive.


5.Modelul arborelui de decizie

```python
# -----------------------------
# Decision Tree Classifier
# -----------------------------

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report

# Crearea modelul Decision Tree
decision = DecisionTreeClassifier()

# Antrenarea modelului pe datele de training
decision.fit(X_train, y_train)

# Prezicerea valorilor pentru setul de test
y_pred_Decision = decision.predict(X_test)

# Calcularea acurateții pe setul de training
decision_train = round(decision.score(X_train, y_train) * 100, 2)

# Calcularea acuratețeii pe setul de test
decision_accuracy = round(accuracy_score(y_pred_Decision, y_test) * 100, 2)

# Calcularea scorului F1 pe setul de test
decision_f1 = round(f1_score(y_pred_Decision, y_test) * 100, 2)

# Afișarea rezultatelor
print("Acuratețea pe setul de training :", decision_train, "%")
print("Acuratețea modelului           :", decision_accuracy, "%")
print("Scorul F1                      :", decision_f1, "%")
print("\033[1m--------------------------------------------------------\033[0m")

# Raport de clasificare (precizie, recall, F1 pentru fiecare clasă)
print("Raport de clasificare:\n", classification_report(y_test, y_pred_Decision))
```

```python
# -----------------------------
# Matricea de confuzie - Decision Tree Classifier
# -----------------------------

plt.rcParams['figure.figsize'] = (10, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 

# Calculare matrice de confuzie
cm = confusion_matrix(y_test, y_pred_Decision)

# Transformarea matriceei într-un DataFrame pentru vizualizare mai clară
cm_df = pd.DataFrame(cm)

# Vizualizare cu Seaborn

sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title("Matricea de Confuzie - Decision Tree")
plt.ylabel("Valori reale")
plt.xlabel("Valori prezise")
plt.show()
```

```python
# -----------------------------
# Curba ROC - Decision Tree Classifier
# -----------------------------

plt.rcParams['figure.figsize'] = (14, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 

# Crearea afișajului curbei ROC pe baza modelului Decision Tree
RocCurveDisplay.from_estimator(
    decision,       # modelul antrenat
    X_test,    # datele de test
    y_test     # valorile reale
)

plt.title("Curba ROC - Decision Tree")

# Afișare figură
plt.show()
```

În comparație cu modelele anterioare (Regresie Logistică și Naive Bayes), arborele de decizie prezintă o performanță similară sau ușor mai bună la nivel de acuratețe și F1-score. Naive Bayes Gaussian are o acuratețe de aproximativ 79.8% și un F1-score pentru clasa pozitivă de 73.6%, în timp ce arborele de decizie atinge o acuratețe de 80.7% și un F1-score de 74.5%, ceea ce îl plasează foarte apropiat, dar ușor superior din perspectiva echilibrului între precizie și recall pentru clasa minoritară. Observația importantă este că arborele are o acuratețe pe training extrem de mare (~99%), semn că tinde să suprainvețe (overfitting), însă pe setul de test performanța rămâne comparabilă cu modelele mai simple, fără un câștig major.

Matricea de confuzie arată modul concret în care modelul clasifică exemplele reale: valorile de pe diagonală (în special 113 și 61) reprezintă predicții corecte pentru clasele 0 și 1, iar valorile din afara diagonalei (15 false pozitive și 24 false negative) indică tipurile de erori ale modelului. Acest tip de analiză este important pentru a vedea dacă modelul ratează mai frecvent clasa pozitivă sau negativă. Curba de tip precizie–recall (alternativă la ROC pentru seturi dezechilibrate) arată cum se modifică precizia și recall-ul când pragul de decizie variază; forma ei și valoarea Average Precision = 0.69 indică faptul că modelul are o capacitate moderată de a separa corect clasa pozitivă, dar nu una excelentă, confirmând ceea ce se vede și în matricea de confuzie.


6.Compararea modelelor si confuzii

```python
# -----------------------------
# Crearea unui DataFrame pentru compararea modelelor
# -----------------------------

import pandas as pd

models = pd.DataFrame({
    'Model': [
        'GaussianNB',           # Numele modelului Gaussian Naive Bayes
        'Decision Tree'         # Numele modelului Decision Tree
    ],
    'Training Accuracy': [
         gaussian_train, decision_train  # Acuratețea pe setul de training
    ],
    'Model f1 Score': [
     gaussian_f1, decision_f1          # Scorul F1 pe setul de test
    ],
    'Model Accuracy Score': [
         gaussian_accuracy, decision_accuracy  # Acuratețea pe setul de test
    ]
})

# Afișarea DataFrame-ul
models
```
